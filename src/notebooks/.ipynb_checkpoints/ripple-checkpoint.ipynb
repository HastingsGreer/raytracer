```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(-1, 1, 50)
y = np.linspace(-1, 1, 50)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2) * 12) / 12
```

```python
plt.imshow(z, interpolation="none")
```

```python
def vec(i, j):
    print x[i, j], y[i, j], z[i, j]

for i in range(50):
    for j in range(50):
        #upper left triangle
        print x[i, j], y[i, j], z[i, j]
        print x[i+1, j], y[i+1, j], z[i+1, j]
```
